```python
def replace_outliers_with_mean_iqr(data, k=1):
    # Calculate the first and third quartiles
    q1 = np.percentile(data, 25)
    q3 = np.percentile(data, 75)

    # Calculate the interquartile range (IQR)
    iqr = q3 - q1

    # Define the lower and upper bounds for outliers
    lower_bound = q1 - k * iqr
    upper_bound = q3 + k * iqr

    # Identify outliers and replace them with the mean
    cleaned_data = np.where((data < lower_bound) | (data > upper_bound), np.mean(data), data)

    return cleaned_data
```

```python
### Augmented Dickey-Fuller Test for stability

import statsmodels.api
from statsmodels.graphics.tsaplots import plot_acf
```

```python
# stability test: Augmented Dikey-Fuller Test
def adf_test(series,title=""):
    print("Augmented Dickey fuller Test :",title)
    result=statsmodels.tsa.stattools.adfuller(series,autolag="AIC")
    adf_stat,p_value,lag_used,observations=result[0:4]
    print("adf_score : ",adf_stat,"\np_value :",p_value,"\nlag_used :",lag_used,"\nobservations :",observations)
    print(result[4])
    
    if result[1] >= 0.05:
        print("Data is not stationary")
    else:
        print("Data is stationary.\nYou can procede VAR")
```

```python
def stat_summ(series):
    plt.plot(series,label="series")
    plt.plot(series.rolling(window=12).mean(),label="series mean")
    plt.plot(series.rolling(window=12).std(),label ="series std")
    plt.legend()
    plt.show()
    plot_acf(series,lags=12)
    plt.show()
```

```python
for i in num_col:
    print(i,"Results")
    adf_test(df[i])
    stat_summ(df[i])
```

# Data Transformation for stationarity

```python
df_tsa=df[num_col]
df_tsa.head()
```

```python
# train and test data split
tests=30
test=df_tsa[:tests]
train=df_tsa[tests:]
```

```python
aic=[]
bic=[]
for i in range(15,40):
    model=statsmodels.tsa.vector_ar.var_model.VAR(train)
    results=model.fit(i)
    aic.append(results.aic)
    bic.append(results.bic)
    print(f"order : {i}")
    print(f"AIC : {results.aic}")
    print(f"BIC : {results.bic}")
aic_bic=pd.DataFrame({"i":range(15,40),
                      "AIC":aic,
                     "BIC":bic})
print(aic_bic)
```

## Time Series Modeling
